```python
raw_data["title"] = raw_data["title"].replace("Mlle", "Miss")
raw_data["title"] = raw_data["title"].replace("Ms", "Miss")
raw_data["title"] = raw_data["title"].replace("Mme", "Mrs") # replace í•¨ìˆ˜ë¥¼ ì‚¬ìš©í•´ì„œ data ë³€ê²½ ê°€ëŠ¥
raw_data["title"]
```

```python
# ë“œë¬¼ê²Œ í‘œí˜„ëœ dataê°’ì€ í•˜ë‚˜ë¡œ í†µí•©í•´ì„œ ì²˜ë¦¬í•´ë²„ë ¤ë¼!
Rare = ['Lady','the Countess','Countess','Capt', 'Master',
        'Col','Don','Dr','Major','Rev','Sir','Jonkheer', 'Dona']

for x in Rare: # forë¬¸ë“¤ í†µí•´ì„œ í•œë²ˆì— title ì»¬ëŸ¼ì— ìžˆëŠ” ê°’(Rareì— ì†í•œ)ì„ Rareë¡œ ë³€ê²½
    raw_data["title"] = raw_data["title"].replace(x, "Rare")

raw_data["title"].unique() # ê°„ëžµí•˜ê²Œ ìˆ˜ì •ê°€ëŠ¥
```

```python
# raw_data.groupby("title").mean() ëª¨ë“  ì»¬ëŸ¼ ë‹¤ ë³´ì´ëŠ” ê·¸ë£¹
raw_data[["title", "survived"]].groupby("title").mean() # titleì„ ì¡°ê±´ìœ¼ë¡œ survived ì»¬ëŸ¼ì„ ì¶œë ¥(mean() í‰ê· ê°’)
```

```python
# ì»´í“¨í„°ëŠ” object íƒ€ìž…ì„ ì¸ì‹í•˜ì§€ ëª»í•œë‹¤. ìˆ«ìžë¡œ ì „ë¶€ ë³€ê²½í•´ì¤˜ì•¼í•œë‹¤.
tmp = ["f"] * 1309 # tmpëŠ” fê°€ 1309ê°œì´ë‹¤.
raw_data2 = raw_data.copy()
raw_data2["sex"] = tmp # raw_data2ì˜ sexì»¬ëŸ¼ì„ ì „ë¶€ tmpë¥¼ ì ìš©í•´ì„œ fë¡œ ë³€ê²½í•´ì¤€ë‹¤.
raw_data2["sex"]
```

```python
tmp2 = []
for x in raw_data["sex"]: # raw_dataì˜ sex ì»¬ëŸ¼ì„ ë³€ê²½í•´ì£¼ëŠ” ê³¼ì •
    if x == "female": # sex ì»¬ëŸ¼ì˜ ê°’ì´ female ì¼ë•Œ
        tmp2.append(0) # ìˆ«ìž 0ìœ¼ë¡œ ë³€ê²½í•´ì¤€ë‹¤
    elif x == "male": # ìœ„ì™€ ê°™ì€ ê³¼ì •
        tmp2.append(1)
    else:
        tmp2.append(np.nan) # numpyë¥¼ í™œìš©í•´ì„œ femaleë„ maleë„ ì•„ë‹Œ ê°’ì€ nanìœ¼ë¡œ ëŒ€ì²´
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 3, random_state = 13) # ê°ì²´ë¥¼ ìƒì„±
# ì–¼ë§ˆë‚˜ ê¹Šê²Œ ë¶„ì„í• ê¸° max_depth, data seedê°’ì€ ê³ ì •ì‹œì¼œì•¼í•œë‹¤. random_state = ê°’ ê³ ì •
tree_clf.fit(X_train, y_train) # fit ê³µë¶€ ì‹œí‚¤ëŠ” í•¨ìˆ˜!
# train dataë¥¼ ê°€ì§€ê³  ê³µë¶€í•œ ì ìˆ˜
tree_clf.score(X_train, y_train) # ì •í™•ë„ê°€ 81% ì •ë„! ì •í™•ë„ í™•ì¸ í•¨ìˆ˜ëŠ” score
```
