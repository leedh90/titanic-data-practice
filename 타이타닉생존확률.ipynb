```python
pd.crosstab(raw_data["title"], raw_data["sex"]) # title에서 성별을 출력해라
# pandas의 crosstab은 첫번째 조건을 기준으로 2번째 값을 출력
# 첫번째 조건값은 여러가지 컬럼으로도 줄 수 있다 [여러가지 조건 컬럼을 넣을 수 있음.]
```

```python
raw_data["title"] = raw_data["title"].replace("Mlle", "Miss")
raw_data["title"] = raw_data["title"].replace("Ms", "Miss")
raw_data["title"] = raw_data["title"].replace("Mme", "Mrs") # replace 함수를 사용해서 data 변경 가능
raw_data["title"]
```

```python
# 드물게 표현된 data값은 하나로 통합해서 처리해버려라!
Rare = ['Lady','the Countess','Countess','Capt', 'Master',
        'Col','Don','Dr','Major','Rev','Sir','Jonkheer', 'Dona']

for x in Rare: # for문들 통해서 한번에 title 컬럼에 있는 값(Rare에 속한)을 Rare로 변경
    raw_data["title"] = raw_data["title"].replace(x, "Rare")

raw_data["title"].unique() # 간략하게 수정가능
```

```python
# raw_data.groupby("title").mean() 모든 컬럼 다 보이는 그룹
raw_data[["title", "survived"]].groupby("title").mean() # title을 조건으로 survived 컬럼을 출력(mean() 평균값)
```

```python
# 컴퓨터는 object 타입을 인식하지 못한다. 숫자로 전부 변경해줘야한다.
tmp = ["f"] * 1309 # tmp는 f가 1309개이다.
raw_data2 = raw_data.copy()
raw_data2["sex"] = tmp # raw_data2의 sex컬럼을 전부 tmp를 적용해서 f로 변경해준다.
raw_data2["sex"]
```

```python
tmp2 = []
for x in raw_data["sex"]: # raw_data의 sex 컬럼을 변경해주는 과정
    if x == "female": # sex 컬럼의 값이 female 일때
        tmp2.append(0) # 숫자 0으로 변경해준다
    elif x == "male": # 위와 같은 과정
        tmp2.append(1)
    else:
        tmp2.append(np.nan) # numpy를 활용해서 female도 male도 아닌 값은 nan으로 대체
```

```python
len(tmp2) # 개수 체크!
```

```python
raw_data["sex"] = tmp2 # raw_data의 sex 컬럼을 tmp2로 변경시킨다.
raw_data["sex"].head()
```

```python
raw_data['survived'] = raw_data['survived'].astype('float') # 각 컬럼의 type을 변경할때는 astype 사용
raw_data['pclass'] = raw_data['pclass'].astype('float')
raw_data['sex'] = raw_data['sex'].astype('float')
raw_data['sibsp'] = raw_data['sibsp'].astype('float')
raw_data['parch'] = raw_data['parch'].astype('float')
raw_data['fare'] = raw_data['fare'].astype('float')
# int 형태의 컬럼들을 전부 float 형태로 변경해준다. 머신러닝은 문자열과 int 형을 인식할 수 없다
# 따라서 형태를 float 형태로 변경해준다.
raw_data.head()
```

```python
raw_data = raw_data[raw_data['age'].notnull()] # 낫널인 값만 뽑아보기!
raw_data = raw_data[raw_data['sibsp'].notnull()]
raw_data = raw_data[raw_data['parch'].notnull()]
raw_data = raw_data[raw_data['fare'].notnull()]
raw_data.info() # 전체 row의 갯수가 같아야 비교할 수 있다.
# 총 갯수는 1309개지만 많지만 거기서 notnull인거 빼고 1045개만 출력
```

```python
train_pre = raw_data[["pclass", "sex", "age", "sibsp", "parch", "fare"]] # 각각의 row갯수가 같은걸로 뽑아준다
# 내가 머신러닝에 사용할 data의 row갯수는 전부 일치하는 컬럼으로만 뽑아줘야한다.
```

```python
train_pre # notnull 값은 빼고 출력하기 때문에 row는 1045지만 
```

```python
# data 준비끝!
# 머신러닝용 라이브러리: sklearn, 사이킬런! 임포트해야한다.
from sklearn.model_selection import train_test_split
# 훈련용 테스트용 data를 만들기 위한 라이브러리 임포트
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_pre, 
                                                    raw_data[["survived"]], 
                                                    test_size = 0.1, 
                                                    random_state = 13)
#데이터 프레임, 맞추어야할 정답, 테스트데이터는 10% 훈련데이터가 90%로 할당, random_state는 일정한 랜덤값생성
#random_state = java의 random(45) 같은 기능 같은 패턴의 난수 발생
# 지금 기준으로 X가 문제 y가 정답이다.
```

```python
X_train = X_train.reset_index()
X_train # 인덱스 정렬을 하고 난수로 발생한 인덱스를 지우면 깔끔하게 정리할 수 있다.
```

```python
X_train = X_train.drop(["index"], axis=1)
# 열을 다 지우기 위해서 axis를 1로 처리해준다. 첫번째 열만 전부 삭제(axis=1)
X_train
```

```python
X_test = X_test.reset_index()
X_test = X_test.drop(["index"], axis=1)

y_train = y_train.reset_index()
y_train = y_train.drop(["index"], axis=1)

y_test = y_test.reset_index()
y_test = y_test.drop(["index"], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier # 트리형식으로 분류하는 분석
```

```python
tree_clf = DecisionTreeClassifier(max_depth = 3, random_state = 13) # 객체를 생성
# 얼마나 깊게 분석할기 max_depth, data seed값은 고정시켜야한다. random_state = 값 고정
tree_clf.fit(X_train, y_train) # fit 공부 시키는 함수!
# train data를 가지고 공부한 점수
tree_clf.score(X_train, y_train) # 정확도가 81% 정도! 정확도 확인 함수는 score
```

```python
from sklearn.metrics import accuracy_score # 채점해주는 라이브러리
y_pred = tree_clf.predict(X_test) # 교육용 데이터가 아닌 데스트용 데이터를 줘서 분석하기 predict 함수
# 예상 답안을 만들기 때문에 리턴값이 있어서 항상 변수에 넣어줘야 한다.
# 정답을 리턴시킨다!
# 정답과 컴퓨터가 푼 답과 얼마나 일치하는지 비교해주기
```

```python
accuracy_score(y_test, y_pred) # accuracy_score(해답, 컴퓨터가 푼답) 정확도를 비교
```
